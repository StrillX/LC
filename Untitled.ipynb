```python
import numpy as np
import math 
import networkx as nx
pontos = [x+1 for x in range(81)]
pontos = np.reshape(pontos,(9,9))
print(pontos)
```

```python
def adjacencia(mat):
    adj=[]
    for y in range(len(mat)):
        for x in range(len(mat[y])):
            for outros in range(x,len(mat[y])):
                adj.append((mat[y][x],mat[y][outros]))
    for y in range(len(mat)):
        for x in range(len(mat[x])):
            for outros in range(y,len(mat)):
                adj.append((mat[y][x],mat[outros][x]))
    for y in range(len(mat)):
        for x in range(len(mat[y])):
            if y % int(math.sqrt(len(mat))) == 0 and x % int(math.sqrt(len(mat))) ==0:
                adj.append((mat[y][x],mat[y+1][x+1]))
                adj.append((mat[y][x],mat[y+1][x+2]))
                adj.append((mat[y][x],mat[y+2][x+1]))
                adj.append((mat[y][x],mat[y+2][x+2]))
            
            elif y % int(math.sqrt(len(mat))) == 0 and x % int(math.sqrt(len(mat))) ==1:
                adj.append((mat[y][x],mat[y+1][x+1]))
                adj.append((mat[y][x],mat[y+1][x-1]))
                adj.append((mat[y][x],mat[y+2][x+1]))
                adj.append((mat[y][x],mat[y+2][x-1]))
            
            elif y % int(math.sqrt(len(mat))) == 0 and x % int(math.sqrt(len(mat))) ==2:
                adj.append((mat[y][x],mat[y+1][x-2]))
                adj.append((mat[y][x],mat[y+1][x-1]))
                adj.append((mat[y][x],mat[y+2][x-2]))
                adj.append((mat[y][x],mat[y+2][x-1]))
                
            elif y % int(math.sqrt(len(mat))) == 1 and x % int(math.sqrt(len(mat))) ==0:
                adj.append((mat[y][x],mat[y+1][x+1]))
                adj.append((mat[y][x],mat[y+1][x+2]))
                
            elif y % int(math.sqrt(len(mat))) == 1 and x % int(math.sqrt(len(mat))) ==1:
                adj.append((mat[y][x],mat[y+1][x+1]))
                adj.append((mat[y][x],mat[y+1][x-1]))
               
            elif y % int(math.sqrt(len(mat))) == 1 and x % int(math.sqrt(len(mat))) ==2:
                adj.append((mat[y][x],mat[y+1][x-2]))
                adj.append((mat[y][x],mat[y+1][x-1]))
        for (x,y) in adj:
            if x==y :
                adj.remove((x,y))
    return adj
                
            
                
        
    
edges=adjacencia(pontos)


```

```python
grafo = nx.Graph(edges)
dic=nx.coloring.greedy_color(grafo, strategy="saturation_largest_first")
 
dic=dict(sorted(dic.items()))
print(dic)
for k in dic:
    dic[k]+=1
    
print(dic)

solvedSudoku = [x for x in dic.values()]
solvedSudoku=np.reshape(solvedSudoku,(int(math.sqrt(len(dic))),int(math.sqrt(len(dic)))))
print(solvedSudoku)
nx.draw_circular(grafo, with_labels= True)
```

```python
def brute_force_color(graph,k):
    color = {}
    # tenta-se preencher o dicionário color com uma coloração válida                  
    if aux(graph,k,color):
        # se for possível então transferem-se as cores
        # para o atributo color dos nós
        for n in graph:
          graph.nodes[n]['color'] = color[n]
        return True
    else:
        return False

# função recursiva que tenta extender o dicionário
# color (incialmente vazio) com uma coloração válida para o grafo
def aux(graph,k,color):
    if len(color) == len(graph):   
        # se o dicionário já tem uma cor para cada nó então
        # testa-se se é válida
        for o in graph:
            for d in graph[o]:
                if color[o] == color[d]:
                    return False
        return True
    # senão escolhe-se o próximo nó a colorir
    v = list(graph.nodes)[len(color)]
    for c in range(k):
        # tenta-se colorir esse nó com uma das cores possíveis
        color[v] = c
        # e recursivamente tenta-se colorir os restantes e
        # obter uma coloração válida
        if aux(graph,k,color): 
            # se for possível termina a pesquisa
            return True
        # se não for possível tenta-se com outra cor
        color.pop(v)
    return False

assert not brute_force_color(petersen,2)
assert brute_force_color(petersen,3)
draw_with_colors(petersen)
print(petersen.nodes.data())
```

```python
def find_optimal_coloring(graph,testa):
    k = 0
    while not testa(graph,k):
        k += 1
    return k

assert find_optimal_coloring(petersen,brute_force_color) == 3
draw_with_colors(petersen)
print(petersen.nodes.data())
```

```python
from ortools.linear_solver import pywraplp
def ip_color(graph,k):
    solver = pywraplp.Solver.CreateSolver('SCIP')
    x={}
    for v in graph:
        for c in range(k):
            x[v,c]= solver.IntVar(0,1,'x[%i][%i]' % (v,c))
    for o,d in graph.edges():
        for c in range(k):
            solver.Add(x[o,c] + x[d,c] <= 1)
    for v in graph:
        solver.Add(sum(x[v,c] for c in range(k))== 1)
    status = solver.Solve()
    if status == pywraplp.Solver.OPTIMAL:
        for v in graph:
            for c in range(k):
                if (round(x[v,c].solution_value==1)):
                    graph.nodes[v]['color'] = c
                return True
    return False
        
assert ip_color(petersen,3)
draw_with_colors(petersen)
print(petersen.nodes.data())
```

```python
def ip_color2(graph,k):
    

    # criar solver
    solver = pywraplp.Solver.CreateSolver('SCIP')
    #criar dicionario de variaveis x{i,j} 
    x = {}
    for i in graph:
        x[i] = {}
        for j in range(k):
            x[i][j] = solver.IntVar(0,1,'x[%i][%i]' % (i,j))
      
    # vertices adjacentes tem cores diferentes
    for o in graph:
        for d in graph[o]:
            for j in range(k):
                  solver.Add(x[o][j] + x[d][j] <= 1)
    
    # cada vertice adjacente tem cores diferentes

    for i in graph:
        solver.Add(sum([x[i][j] for j in range(k)]) == 1)  # ou solver.Add(sum(list(x[i].values())) == 1) 

    # invocar solver e colorir o grafo

    stat = solver.Solve()
    if stat == pywraplp.Solver.OPTIMAL:

      # colorir

        for i in graph:
            for j in range(k):
                if round(x[i][j].solution_value()) == 1:
                    graph.nodes[i]['color'] = j+1
        return True
    else:
        return False
```
